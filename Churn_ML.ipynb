```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.impute import KNNImputer
```

```python
data = pd.read_csv('E Commerce Dataset.csv')
data
```

```python
data.describe(include='all')
```

### Extracting the numerical and Categorical columns

```python
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
numerical_cols
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
categorical_cols
```

#### Dropping 'CustomerID' from numerical columns as it's an identifier, not a feature

```python
numerical_cols.remove("CustomerID")
numerical_cols
```

### Handling missing values

```python
missing_value_count = data.isnull().sum()
missing_value_count
```

### Imputting the missing values in the numerical columns using KNN method

```python
## Initialising the imputer

imputer = KNNImputer(n_neighbors=5)
```

```python
# Creating a copy of the numerical data to avoid modifying the original DataFrame
numerical_data_to_impute = data[numerical_cols].copy()

# Fitting and transforming the numerical data to impute the missing values
imputed_data = imputer.fit_transform(numerical_data_to_impute)

# Convert the imputed numpy array back to a DataFrame with original column names
imputed_numerical_data = pd.DataFrame(imputed_data, columns=numerical_cols)
imputed_numerical_data
```

### Imputting the missing values in the categorical columns using One-Hot Encoding

```python
categorical_data = pd.get_dummies(data[categorical_cols])
categorical_data
```

### Combining the processed data

```python
final_data = pd.concat([imputed_numerical_data, categorical_data], axis=1)
final_data
```

```python
# Displaying a summary of the final DataFrame to confirm there are no missing values
final_data.info()
```

```python
# Display a descriptive summary of the numerical columns
print("\nDescriptive statistics for numerical columns:")
print(final_data.describe())
```
